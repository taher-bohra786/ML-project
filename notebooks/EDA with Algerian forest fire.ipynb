Data Cleaning

```python
#Missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is convered intot two sets based on Regio from 122th index , we can make a new column based on the Region

1. "Bejaia Region Dataset
2. "Sidi-Bel Abbes Region Dataset

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

Check for the missing values

```python
df.isnull().sum()
```

Removing the null values

```python
df = df.dropna().reset_index(drop = True)  # 2 Null values removed so ow it is 246 to 244
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] #We will remove this too
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]  #Different data has taken the place at this position
```

```python
df.columns
```

```python
#Fix spaces in column names
df.columns = [col.replace(' ', '') for col in df.columns]
# or 
# df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()   #All columns are in object
```

Changes the required columns as integer data type

```python
df.columns
#Here 'day', 'month', 'year', 'Temperature', 'RH', 'Ws' this columns are all have numeric values
```

```python
#Converting those columns into integer
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info() 
```

Changing the other columns into foating data type 

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
objects
```

```python
df['Classes'].head()
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Saving the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv', index = False)
```

VISUALIZATION

```python
#Density plot for all features
sns.set_style('darkgrid')   #Set seaborn style
df_copy.hist(bins=50 , figsize = (20,15))
plt.show()
```

```python
#Percentage of the pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
#plotting pie chart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage, labels = classlabels , autopct='%1.1f%%')
plt.title("Pie chart of the classes")
plt.show()
```

```python
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize = (13,6))
sns.set_style('whitegrid')
sns.countplot(x='month' , hue = 'Classes' , data=df)
plt.ylabel('Number of fires' , weight='bold')
plt.xlabel('Months' , weight='bold')
plt.title('Fire analysis of Brjaia Regions',weight='bold')
```
